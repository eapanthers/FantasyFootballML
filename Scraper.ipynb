```python
import csv
from datetime import datetime
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2
from bs4 import BeautifulSoup
with urllib.request.urlopen("https://www.pro-football-reference.com/years/2017/fantasy.htm") as url:
    page = url.read()
    
    
soup = BeautifulSoup(page,"lxml")
table = soup.find('table')

list_of_rows = []
for row in table.findAll('tr'):
    list_of_cells = []
    for row1 in row.findAll(["th","td"]):
        text1 = row1.text
        if text1 == 'WR':
            for cell in row.findAll(["th","td"]):
                text = cell.text
                list_of_cells.append(text)
    list_of_rows.append(list_of_cells)

list_of_rows = list(filter(None, list_of_rows))

for item in list_of_rows:
    print(' '.join(item))
    
with open('wrStats.csv', 'a') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(list_of_rows)
```
